```python
def getODBconn(user, hostname, db, pw = None):
    if pw == None:
        print("Enter password for " + user + "@" + db + ": ")
        pw = getpass.getpass()
    connection=oracledb.connect(user=user, password=pw, host=hostname, service_name=db)
    engine = create_engine('oracle+oracledb://', creator=lambda: connection)
    return engine

###

# Function to run a SQL query and input result into dataframe

def run_query(query):
    eng = getODBconn('PARSCA10_SQL', 'AIX47', 'BCDWHPRD')
    cnx = eng.connect()
    qry = cnx.execute(query)
    qry_df = pd.DataFrame(qry.fetchall())
    qry_df.columns = qry.keys()
    return qry_df

###

# Function to extract the first and last word from a string
# Example: df['first_and_last_words'] = df['col_of_messy_strings'].apply(first_and_last_word)

def clean_rep_names(s):
    s = s.replace(".", "")
    words = s.split(',')
    if len(words) > 1:
        last_name = words[0].strip()
        first_mid_name_words = words[-1].strip().split(' ')
        if ((len(first_mid_name_words) > 1) & (len(first_mid_name_words[-1])==1)):
            first_name = first_mid_name_words[0]
        else:
            first_name = ' '.join(first_mid_name_words)
        name = first_name + " " + last_name
        return name.title()
    elif words:  # If there is only one word
        return words[0]
    else:  # If the string is empty
        return ''

###    

# Function to find the string that best matches another using rapidfuzz package

def find_best_match(name, choices, scorer=fuzz.token_sort_ratio):
    match, score, _ = process.extractOne(name, choices, scorer=scorer)
    return match, score

def clean_rep_rank_df(dataframe):
    dataframe['Cleaned_Rep_Name'] = dataframe['Rep'].apply(clean_rep_names)
    
    # Create a new DataFrame to store matches and scores
    matches = []

    for name in df['Rep_Name_in_Other'].unique():
        match, score = find_best_match(name, dataframe['Cleaned_Rep_Name'])
        firstlast = name.split()
        match2, score2 = find_best_match(firstlast[-1], dataframe['Cleaned_Rep_Name'])
        matches.append({'Rep_Name_in_Other': name, 'Cleaned_Rep_Name': match, 'score': score, '2nd Match': match2, '2nd Score': score2})

    matches_df = pd.DataFrame(matches)
    matches_df.head(10)

    matches_df['Rep_Name_in_Other'] = matches_df['2nd Match'].where(matches_df['score'] < matches_df['2nd Score'], matches_df['Rep_Name_in_Other'])

#     Set a score threshold for acceptable matches if needed 
    score_threshold = 80
    matches_df = matches_df[matches_df['score'] >= score_threshold]

    matches_df = matches_df.drop(['2nd Match', '2nd Score'], axis=1)
#     # matches_df

    dataframe2 = pd.merge(dataframe, matches_df, on='Cleaned_Rep_Name', how='left').drop('score', axis=1)
    dataframe2 = dataframe2[~dataframe2['Rep_Name_in_Other'].isna()]
    return dataframe2

def wrangle_master_df(rep_df, df_to_filter):
    valid_names = set(rep_df['Cleaned_Rep_Name']).union(rep_df['Rep_Name_in_Other'])
    df_to_filter = df_to_filter[df_to_filter['Rep_Name_in_Other'].isin(valid_names)]

    df_to_filter = pd.merge(df_to_filter, 
                              rep_df[['Ty', 'Performance Category', 'Cleaned_Rep_Name', 'Rep_Name_in_Other']], 
                              left_on='Rep_Name_in_Other', right_on='Rep_Name_in_Other')
    
    # Convert 'Performance Category' to ordinal encoding
    ordinal_mapping = {'low': 1, 'average': 2, 'high': 3}
    df_to_filter['Performance_Category_Ordinal'] = df_to_filter['Performance Category'].map(ordinal_mapping)
    return df_to_filter
```

```python
def casket_quantity_corr(dataframe):
    dataframe2 = dataframe[(dataframe['quota_cat_cd4_bl05'] == 'BUR') & (dataframe['srp10'] == 'CSKT')]
    
    # Summarize the average quantity ordered per rep by rep performance category
    summary = dataframe2.groupby(['Cleaned_Rep_Name', 'Performance_Category_Ordinal'])['units_shipped'].sum().reset_index()
    average_summary = summary.groupby('Performance_Category_Ordinal')['units_shipped'].mean().reset_index()
    average_summary.columns = ['Performance_Category_Ordinal', 'avg_units_shipped_per_rep']
    average_summary

    # Bar plot to visualize the average quantity ordered per rep by rep performance category
    plt.figure(figsize=(6, 4))
    sns.barplot(x='Performance_Category_Ordinal', y='units_shipped', data=summary)
    plt.title('Avg Quantity of Caskets Ordered per Rep by Rep Performance Category')
    plt.xlabel('Rep Performance Category')
    plt.ylabel('Average Quantity Ordered per Rep')
    plt.show()


    # Calculate Spearman correlation between rep performance category and average quantity ordered per rep
    corr, p_value = spearmanr(summary['Performance_Category_Ordinal'], summary['units_shipped'])

    print(f"Spearman Correlation: {corr}")
    print(f"P-value: {p_value}")
    
def newpointe_casket_corr(dataframe):
    dataframe2 = dataframe[(dataframe['quota_cat_cd4_bl05'] == 'BUR') & (dataframe['srp10'] == 'CSKT')]
    newpointe_df = dataframe2[(dataframe2['primary_product_type'] == 'NewPointe')]
    newpointe_summary = newpointe_df.groupby(['Cleaned_Rep_Name', 'Performance_Category_Ordinal'])['units_shipped'].sum().reset_index()
    newpointe_summary['np_units_shipped'] = newpointe_summary['units_shipped']
    newpointe_summary = newpointe_summary.drop('units_shipped', axis=1)

    # Summarize the average quantity ordered per rep by rep performance category
    summary = dataframe2.groupby(['Cleaned_Rep_Name', 'Performance_Category_Ordinal'])['units_shipped'].sum().reset_index()
    
    newpointe_corr_df = pd.merge(newpointe_summary, summary, on=['Cleaned_Rep_Name', 'Performance_Category_Ordinal'])
    newpointe_corr_df['pct_newpointe'] = newpointe_corr_df['np_units_shipped'] / newpointe_corr_df['units_shipped']
    
#     average_summary = summary.groupby('Performance_Category_Ordinal')['units_shipped'].mean().reset_index()
#     average_summary.columns = ['Performance_Category_Ordinal', 'avg_units_shipped_per_rep']
    print(newpointe_corr_df['pct_newpointe'].max(), newpointe_corr_df['pct_newpointe'].min())

    # Bar plot to visualize the average quantity ordered per rep by rep performance category
    plt.figure(figsize=(6, 4))
    sns.barplot(x='Performance_Category_Ordinal', y='pct_newpointe', data=newpointe_corr_df)
    plt.title('Avg % NewPointe caskets per Rep by Performance Category')
    plt.xlabel('Rep Performance Category')
    plt.ylabel('Average Quantity NewPointe Caskets Ordered')
    plt.show()


    # Calculate Spearman correlation between rep performance category and average quantity ordered per rep
    corr, p_value = spearmanr(newpointe_corr_df['Performance_Category_Ordinal'], newpointe_corr_df['np_units_shipped'])

    print(f"Spearman Correlation: {corr}")
#     print(f"P-value: {p_value}")
    
def urn_quantity_corr(dataframe):
    dataframe2 = dataframe[(dataframe['level_1_descr'] == 'Cremation Options') & (dataframe['srp10'] == 'URN')]
    
    # Summarize the average quantity ordered per rep by rep performance category
    summary = dataframe2.groupby(['Cleaned_Rep_Name', 'Performance_Category_Ordinal'])['units_shipped'].sum().reset_index()
    average_summary = summary.groupby('Performance_Category_Ordinal')['units_shipped'].mean().reset_index()
    average_summary.columns = ['Performance_Category_Ordinal', 'avg_units_shipped_per_rep']
    average_summary

    # Bar plot to visualize the average quantity ordered per rep by rep performance category
    plt.figure(figsize=(6, 4))
    sns.barplot(x='Performance_Category_Ordinal', y='units_shipped', data=summary)
    plt.title('Avg Quantity of Urns Ordered per Rep by Rep Performance Category')
    plt.xlabel('Rep Performance Category')
    plt.ylabel('Average Quantity Ordered per Rep')
    plt.show()


    # Calculate Spearman correlation between rep performance category and average quantity ordered per rep
    corr, p_value = spearmanr(summary['Performance_Category_Ordinal'], summary['units_shipped'])

    print(f"Spearman Correlation: {corr}")
    print(f"P-value: {p_value}")
```
